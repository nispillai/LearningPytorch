```python
import pandas as pd
import numpy as np

poultryFolder = "NACA-USDA_Poultry/"
labFile = poultryFolder + "Pastured_Poultry_Summary_DB_03162020.xlsx"

lab = pd.read_excel(labFile)
```

```python
lab
```

```python
lab = lab[lab['SampleType'] == 'Feces']
lab
```

```python
preVariables = ['pH','EC','Moisture','TotalC','TotalN','CNRatio','Ca','Cd','Cr','Cu','Fe','K','Mg','Mn','Mo','Na','Ni','P','Pb','Zn']
targetVariables = ['Salmonella']
X = lab[preVariables]
Y = lab[targetVariables]
X
```

```python
Y = Y.values.tolist()
```

```python
YY = [0 if yy[0] == "+" else 1 for yy in Y]
YY
```

```python
X1 = np.array(X.values.tolist())
X1.shape, len(YY)
```

```python
edge1 = [1, 2, 4, 5]
edge2 = [11,67,93,102]
edge_index = []
edge_index.append(edge1)
edge_index.append(edge2) 
print(edge_index)
```

```python
from torch_geometric.data import Data
from numpy import dot
from numpy.linalg import norm
import torch
import torch.nn.functional as F
from torch_geometric.nn import GCNConv

xGraph = torch.tensor(X1, dtype=torch.float32) 
yGraph = torch.tensor(np.asarray(YY), dtype=torch.float32)
edge_index = torch.tensor(edge_index, dtype = torch.int64)
trainD = Data(x=xGraph, edge_index=edge_index.contiguous(), y=yGraph)

data = trainD

print(f'Number of nodes: {data.num_nodes}')
print(f'Number of edges: {data.num_edges}')
print(f'Average node degree: {data.num_edges / data.num_nodes:.2f}')
print(f'Has isolated nodes: {data.has_isolated_nodes()}')
print(f'Has self-loops: {data.has_self_loops()}')
print(f'Is undirected: {data.is_undirected()}')
data
```

```python
class GCN(torch.nn.Module):
    def __init__(self):
        super().__init__()

        self.conv1 = GCNConv(data.x.shape[1], 10)
        self.conv2 = GCNConv(10, 1)
        self.sigm = torch.nn.Sigmoid()
        
    def forward(self, data):
        x, edge_index = data.x, data.edge_index

        # notice that for graph convolution layer, we need to fed in 
        # data, as well as edge_index
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, training=self.training)
        x = self.conv2(x, edge_index)             
        return self.sigm(x)
 
model = GCN()
optimizer = torch.optim.Adam(model.parameters(), lr=0.0001, weight_decay=5e-4)
criterion = torch.nn.BCELoss()

# Initializing Model

model.train()


loss_list = []

for epoch in range(50):
    optimizer.zero_grad()
    out = model(data)
    loss = criterion(out, data.y.type(torch.float).unsqueeze(1))
#    loss = criterion(torch.exp(out)[:,0], data.y)
    loss_list.append(loss)
    loss.backward()
    optimizer.step()
```

```python
from sklearn.metrics import *
def calculate_metrics(pred, target, threshold=0.5):
    pred = np.array(pred > threshold, dtype=float)
    return {'micro/precision': precision_score(y_true=target, y_pred=pred, average='micro'),
            'micro/recall': recall_score(y_true=target, y_pred=pred, average='micro'),
            'micro/f1': f1_score(y_true=target, y_pred=pred, average='micro'),
            'macro/precision': precision_score(y_true=target, y_pred=pred, average='macro'),
            'macro/recall': recall_score(y_true=target, y_pred=pred, average='macro'),
            'macro/f1': f1_score(y_true=target, y_pred=pred, average='macro'),
            }


model.eval()
model_result = []

with torch.no_grad():
    model_batch_result = model(data)
    model_result.extend(model_batch_result.cpu().numpy())
    
result = calculate_metrics(np.array(model_result), np.array(data.y))
print(result)
```
